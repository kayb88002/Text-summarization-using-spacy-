```python
import spacy
from spacy import displacy
from googlesearch import search
from newspaper import Article
import sklearn
from spacy.lang.en.stop_words import STOP_WORDS

corpus = ""
```

```python
query=input("Enter your query: ")

for i in search(query,tld="com",num=10,stop=20,pause=2.0):
    print(i)
    article=Article(i)
    article.download()
    article.parse()
    print(article.text)
    corpus = corpus+" "+article.text
```

```python
nlp = spacy.load('en_core_web_lg')

docx = nlp(corpus)
```

```python
mytokens = [token.text for token in docx]
```

```python
stopwords = list(STOP_WORDS)
```

```python
word_freq={}
for word in docx:
        if word.text not in stopwords:
            if word.text not in word_freq.keys():
                word_freq[word.text]=1;
                
            else:
                word_freq[word.text]+=1;
```

```python
max_freq=max(word_freq.values())

for word in word_freq:
    word_freq[word]=(word_freq[word]/max_freq)
```

```python
sentence_list = [ sentence for sentence in docx.sents]
```

```python
[w.text.lower() for t in sentence_list for w in t ]
```

```python
#Calculating Sentence Scores-->
sentence_scores={}
for sentence in sentence_list:
    for word in sentence:
        if word.text in word_freq.keys():
            if len(sentence.text.split(' ')) < 30:
                if sentence not in sentence_scores.keys():
                    sentence_scores[sentence] = word_freq[word.text]
                else:
                    sentence_scores[sentence] += word_freq[word.text]
```

```python
from heapq import nlargest
sum_sentences = nlargest(20, sentence_scores, key=sentence_scores.get)
```

```python
final_sentences = [ w.text for w in sum_sentences ]
```

```python
summary = ' '.join(final_sentences)
```

```python
#remove duplicates
def unique_list(summary):
    ulist = []
    [ulist.append(x) for x in summary if x not in ulist]
    return ulist
print(' '.join(unique_list(summary.split())))
```
